```python
import pandas as pd

input_file = "input\\Sample - Superstore.csv"
output_file = "output\\Superstore_FINAL_FIXED.csv"

# Read very loosely (ignore structure issues)
df = pd.read_csv(
    input_file,
    encoding='latin1',
    engine='python',
    on_bad_lines='skip'
)

# Remove problematic quotes completely
df = df.replace('"', '', regex=True)

# Save without quotes
df.to_csv(
    output_file,
    index=False,
    encoding='utf-8',
    quoting=1  # minimal quoting
)

print("✅ FINAL CLEANED FILE CREATED")
```
